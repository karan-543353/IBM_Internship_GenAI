```python
file = open("students.txt", "w+")
file.write("harnoor : 23 \n nidhish : 56 \n karan : 54 \n simran : 43 \n jaskaran : 54 ")
file.close()

file = open("students.txt","r")
cont = file.read()
print(cont)
file.close()

file = open("students.txt", "r")
pass_count = 0
fail_count = 0
for l in file:
    name, marks = (l.split(":"))
    marks = int(marks.strip())
    name = name.strip()
    if marks > 45:
        pass_count += 1
        print(name, "is pass with marks : ", marks)
    else:
        fail_count += 1
        print(name, " is fail with marks : ", marks)

file.close()

print('pass :', pass_count)
print('fail :', fail_count)

```
